```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv', encoding=encoding)
```

```python
df.shape
```

```python
df.columns
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df[['age','time_spent','income']] = df[['age','time_spent','income']].astype('int')
```

```python
df['time_spent'].dtypes
```

```python
df.describe()
```

**EDA**

```python
df.columns
```

```python
#time spent vs paltform
tp=df.groupby(['platform'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False)
ax=sns.barplot(x='platform',y='time_spent',data=tp)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
tp=df.groupby(['platform','location'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False)
ax=sns.barplot(x='platform',y='time_spent',data=tp,hue='location')
for bars in ax.containers:
  ax.bar_label(bars)
```

**From above graph we can conclude that most time spent is on instagram platform by the users in Australia followed by UK,US**

**PROFESSION**

```python
profession=df.groupby(['profession'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='profession',y='time_spent',data=profession)
```

```python
profession=df.groupby(['profession','platform'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='profession',y='time_spent',data=profession, hue='platform')
```

**From above graph we can see that Market Managers spent most of the time on social media while software engineers spent spent most hours on instagram**

**INCOME**

```python
income=df.groupby(['income'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='income',y='time_spent',data=income)
```

**From above graph we can see people having salary more than 19000 spent most of the time on social media**

**AGE**

```python
df.columns
```

```python
age=df.groupby(['age'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='age',y='time_spent',data=age)
```

**From above graph we can see people of age above 40 tend to spent more time on social media**

**HOME OWNER**

```python
owner=df.groupby(['isHomeOwner'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False).head(10)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='isHomeOwner',y='time_spent',data=owner)
```

**From above graph we can see home owners spent most time on social media.**

**OWN CAR**

```python
car_owner=df.groupby(['Owns_Car'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False).head(10)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='Owns_Car',y='time_spent',data=car_owner)
```

**From above graph we can see car owners spent most time on social media.**

**INDEBT**

```python
indebt=df.groupby(['indebt'],as_index=False)['time_spent'].sum().sort_values(by='time_spent',ascending=False).head(10)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x='indebt',y='time_spent',data=indebt)
```

**From above graph we can see both indebt and not indebt spent equal amount of time on social media.**

**CONCLUSION:**
**Market Manager professionals from Australia above age of 40 spent most of the time on instagram with income more than 19000 and also owns a car and a house. **
